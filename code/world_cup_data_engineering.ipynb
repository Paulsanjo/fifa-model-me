```python
for index, row in df.iterrows():
    home_Elo_Df =  current_elo_df[(current_elo_df["Team"] == row["Squad"]) & (current_elo_df["Date"] <= row["Date"])]
    away_Elo_Df =  current_elo_df[(current_elo_df["Team"] == row["Opponent"]) & (current_elo_df["Date"] <= row["Date"])]
    dr = home_Elo_Df.Elo.values[0]+100 -away_Elo_Df.Elo.values[0]
    e = round(1/(10**(-dr/400) + 1),3)
    r = [1,0.5,0]
    h_elo = 0.0
    a_elo = 0.0
    if row["FTR"]=="W":
        p_h = 20*(r[0]-e)
        p_a = 20*(r[-1]-e)
        h_elo = home_Elo_Df.Elo.values[0] + p_h
        a_elo = away_Elo_Df.Elo.values[0] + p_a
    elif row["FTR"] == "D":
        p_h = 20*(r[1]-e)
        p_a = 20*(r[1]-e)
        h_elo = home_Elo_Df.Elo.values[0] + p_h
        a_elo = away_Elo_Df.Elo.values[0] + p_a
    else:
        p_h = 20*(r[-1]-e)
        p_a = 20*(r[0]-e)
        h_elo = home_Elo_Df.Elo.values[0] + p_h
        a_elo = away_Elo_Df.Elo.values[0] + p_a
    
    df.at[index, 'Team1_Elo']=h_elo
    df.at[index, 'Team2_Elo']=a_elo
    current_elo_df.at[home_Elo_Df.index[0], 'Elo']=h_elo
    current_elo_df.at[home_Elo_Df.index[0], 'Date']=row["Date"]
    current_elo_df.at[away_Elo_Df.index[0], 'Elo']=a_elo
    current_elo_df.at[away_Elo_Df.index[0], 'Date']=row["Date"]
```
