```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("worldcupmatchnew.csv")
data
```

```python
len(data)
```

```python
data["Date"] = pd.to_datetime(data["Date"])
```

```python
#df.groupby()
```

```python
data = data.sort_values(by = ["Date"],ascending = [True]).reset_index()
```

```python
data.drop(["index"],axis=1,inplace = True)
```

```python
data = data.fillna(0)
```

```python
data
```

```python
data.columns
```

```python
COLS = ["Date","Comp","Venue","Squad","Opponent","Ast","PKatt","SoT","CrdY","CrdR","Touches","Tkl","Int","Blocks","Cmp%","Succ","Fls","Off","PKwon","Defensive Tkl%","Defensive Sh","Defensive Pass","Defensive Tkl+Int","Defensive Clr","FTR","Squad_Goals","Opponent_Goals"]
```

```python
data = data[COLS]
```

```python
data
```

```python
data = data.groupby(["Date","Venue","Squad","Opponent","FTR"]).mean()
```

```python
data
```

```python
data=data.reset_index()
```
